# Змістовний модуль 1: Оптика

## Геометрична оптика та Механіка

Геометрична оптика це наближена теорія, яка виникає при нехтуванні масштабів зміни електромагнітної енергії, яка росповсюджується у формі електроманітних хвиль. Отже довжина хвиль вважається нескінченно малою у порівнянні із іншими характерними масштабами (розміри середовища, та масштаби зміни його оптичних властивостей). Основною характеристикою середовища з точки зору геометричної оптики є показник заломлення $n$ який визначає швидкость розповсюдження хвильового фронту $c/n$ у порівнянні із швидкістю світла в вакуумі $c$.


### Оптико-механічна аналогія.
В цьому розділі викладається зв'язок геометричної задачи
із задачами механики і геометричної оптики. 

Світлові явища, як відомо, не є механістичними за своєю природою. Початково сам Ньютон вважав прямолінійність розповсюдження світла проявом механічного руху, представляючи його як потік пружніх малих кульок. У подальшому така корпускулярна теорія була змінена на хвильову (Гюйгенс, Юнг, Френель). Однак певною мірою, Гамільтон реабілітував ідею
Ньютона, застосував ідеї геометричної оптики у механіці, а також втановив
точну відповідність між ними як граничними випадками хвильової теорії. Далі ця ідея взаємопов'язанності хвильових та корпускулярних ефектів виникла знову на початку ХХ століття у вигляді квантової теорії взаємодії матерії та світла. 

З механіки відомо, що рух матеріальної точки під дією жорстких ідеальних зв'язків, які обмежують свободу руху, може бутий представлений як вільний рух частинки по викривленій поверхні. Ця поверхня визначається відповідними  зв'яззками (такі зв'язки називаються голономними або інтегрованими). Рух відбувається вздовж ліній найкоротшої довжини на поверхні - геодезичних ліній. Це також використовують і у зворотньому напрямку як принцип звільнення зв'язків, коли їх замінюють на сили реакції. Тут ми стикаємось із __зв'язком геометрії та механіки__, а більш загально, __геометрії та фізики__. Вкрай важливим є встановлення зв'язків між різними, на перший поляд, ніяк не пов'язаними задачами. В даному випадку це геометрична задача про лінії найкоротшої довжини (геодезична) на поверхні і фізична задача про рух точки під дією сил - траєкторія реального руху. Цікаво що цей звязок може бути виражений в рамках єдиного __принципа екстремальної дії__ математичним апаратом якої є __варіаційне числення__.

#### Принцип Ферма.

Відомо, що в однорідному середовищі світло розповсюджується прямолінійно із постійною швидкіcтю. З точки зору геометрії Евкліда пряма є лінією найкоротшої довжини. У неоднорідному середовищі локальна швидкість світла може залежити від точки 

$$c_{\text{loc}} = \frac{c}{n}$$ 

і визначається показником заломлення $n$ (оптичною густиною).  Тому такий __експериментальний факт__ може бути узагальнений наступним <a id='fermatprncp'> __принципом Ферма__</a>:

__Світло розповсюджується локально від точки А до точки В по шляху ${\rm Path}[A,B]$ який відповідає екстремуму функціоналу часу, наприклад мінімуму. Загальна ситуація відповідає, взагалі кажучи, стаціонарному значенню цього функціоналу (максимум чи точка перегину).__

Середовище можно розділити на невеличкі об'єми, в яких показник заломлення можна вважати сталим. Тоді обчислення руху проміня зводиться до задачі про заломлення світла на границі розділу двох середовищ з показниками заломлення $n_{1}, n_{2}$. Цю задачу вже легко розв'язати (див. Рис)

<img src="./figs/fig_animfermat.gif" alt="Image" width="400">

Виходячи із наведеного вище принципа мінімума часу розповсюдження ($c=1$):

$$
T = n_1\,\sqrt{h^2_{1}+x^2}+n_2\,\sqrt{h^2_{2}+(1-x)^2}
$$

лекго знаходимо, що:

$$ \label{eq_opt12}
n_1\,\frac{x}{\sqrt{h^2_{1}+x^2}} = n_2\,\frac{1-x}{\sqrt{h^2_{1}+(1-x)^2}} \Rightarrow
 \frac{\sin{\alpha_1}}{\sin{\alpha_2}} = \frac{n_2}{n_1}
$$
Тобто __промінь заломлюється у напрямку збільшення показника заломлення (оптичної густини)__. При переході від менш щільного до більш густого середовища кут заломлення менший від кута падіння, та навпаки.

```python
x,h1,h2,n1,n2 = symbols('x h1 h2 n1 n2')
f1=n1*sqrt(x**2+h1**2)+n2*sqrt((1-x)**2+h2**2)
diff(f1,x)
```

Це є т.з __закон заломлення Снеліуса__. В загальному випадку траекторія проміня  відповідає стаціонарному значенню функціонала часу:

$$\label{eq_timefunc}
\left. \delta T_{AB}\right|_{real} = 0\,,\,\,\, T_{AB} = \int\limits_{A}^{B}\frac{n\,d s}{c}
$$

Величина 

$$
d s_{\text{opt} } = n\,d s
$$

називається оптичною довжиною. Іншими словами, світло локально розповсюджується вздовж найкоротшого оптичного шляху.<br>

### Принцип Мопертюї-Ейлера-Лагранжа
Тепер розглянемо задачу про "механічне" заломлення.

<!--  -->
<img src="./figs/fig_mech_prelom.png" alt="Image" width="300">

Частинка рухається з області з постійним потенціалом $\Pi _{1}$ до області
з іншим значенням потенціалу $\Pi _{2}$. Виначити механічний аналог закона
заломлення траєкторії при проходжені границі зміни потенціальної енергії.<br>

__Розв'язок :__ Оскільки сила діє тільки по нормалі до поверхні розділу тангенціальна компонента імпульса зберігається (повна енергія також), тому кути падіння та заломлення задовльняють умові:

$$\label{eq_mech12}
v_{1}\,\sin{\alpha_1} = v_{2}\,\sin{\alpha_2}
$$

Порівнюючи \eqref{eq_mech12} з \eqref{eq_opt12}, маємо, що механічним аналогом принципу Ферма, буде наступний:

__Серед всіх можливих траєкторії руху матеріальної частинки з певною енергією у полі потенціальної сили від точки А до точки В реальна траєкторія відповідає стаціонарному значенню (мінімум, максимум чи перегин) функціонала Дії__

$$\label{eq_act0}
S = \int\limits^{B}_{A}\,p\,ds\\
$$

відповідно

$$\label{sp}
\nabla S = \mathbf{p}\,, \,\,\,\,\, p = \sqrt{2m(E-\Pi)}
$$

Функціонал \eqref{eq_act0} називається скороченою дією і відіграє важливу роль у фізиці.

Звідси витікає оптико-механічна аналогія: рух точки у потенціальному полі з певним значенням повноъ енергії $E>0$ еквівалентний руху проміня світла у оптично неоднорідному середовищі з показником заломлення:

$$
n = \sqrt{1-\frac{\Pi}{E}}
$$

Зазначимо, що в класичній механіці немає обмеження на швидкість руха тому той факт що $n<1$ не є протиріччям в такому підході.

Приклад: траєкторії руху у однорідному полі тяжіння (вільне падіння) як відомо є параболами чи вертикальними прямими. Тому промені у середовищі з показником заломлення:

$$
n(z) = \sqrt{1-\frac{z}{z_0}}\,,\,\,\,\,\, 0\le z \le z0
$$

матимуть таку ж саму форму (параметр $z_0$ визначає набільшу висоту підйому із заданою енергією $E = m\,g\,z_0$).

Отримана аналогія між задачами геометричної оптики та механіки дозволяє встановити відповідність між рівняннями руху задля повноти картини. Рівняння Ньютона:

$$
\label{eq_newton0} m\,\frac{d\,\mathbf{v}}{d\,t} = - \nabla\, \Pi(\mathbf{r})
$$

вводячи одиничний вектор дотичний до траєкторії $\boldsymbol{\tau} = \mathbf{v}/v$, запишем у вигляді:

$$
\label{eq_newton1} v^{2}\frac{d\,\boldsymbol{\tau}}{d\, s} = -
\frac{1}{m}{\nabla} \Pi(\mathbf{r}) - \frac{1}{2}\frac{d\,v^{2}}{d\,s}\boldsymbol{\tau}
$$

или

$$
\frac{d\,\boldsymbol{\tau}}{d\,s} = \frac{\mathbf{F} - (\mathbf{F} \cdot
\boldsymbol{\tau})\,\boldsymbol{\tau}}{2\,(E - \Pi)}\,.
$$

Враховуючи \eqref{eq_newton0}, остаточно отримаєм натуральне рівняння проміня:

$$\label{eqnorm_ray}
 \frac{d\,\boldsymbol{\tau}}{d\,s} = k\,\mathbf{n} =
 \frac{1}{n}\left( \nabla n -
 \frac{d\,n}{d\,s}\boldsymbol{\tau}\right)\,,\\
 k = \frac{\mathbf{n}\cdot\nabla n}{n}
$$
де $k$ - кривизна.<br>

### Принцип стаціонарної дії в механіці
Рівняння руху механічної системи можн вивести з варіаційного принципу стаціонарної дії. Дійсна траєкторія механічного руху між двома точками $A,B$ в фіксовані моменти часу  $t_{A}, t_{B}$  є стаціонарною точкою функціонала дії Лагранжа:
$$
S_{L} = \int\limits_{A}^{B} L(x,\dot{x},t)\,d\,t\,,\,\\ 
\delta S_{L} = 0 \,\,\Rightarrow 
\frac{\delta\,L}{\delta\,x^{i}} = 
\frac{\partial L(x,\dot{x},t)}{\partial x^{i}} -\frac{d}{d\,t}\left(\frac{\partial L(x,\dot{x},t)}{\partial \dot{x}^{i}}\right) = 0
$$
де функція Лагранжа має вигляд різниці кінетичної та потенціальної енергій:
$$
L(x,\dot{x},t) = T - U = m\frac{\mathbf{v}^2}{2} - U(\mathbf{r}) \\
$$
імпульс може бути виражений як функція координат та швидкостей завдяки співвідношенню:
$$
\mathbf{p} = \frac{\partial L}{\partial \mathbf{v}}
$$

Неважко перевірити, що рівняння Ейлера-Лагранжа співпадають з рівнянням Ньютона та функціонал дії може бути представлений у формі Гамільтона:
$$
S_{L} = \int\limits_{A}^{B} L(x,\dot{x},t)\,d\,t = \int\limits_{A}^{B} \mathbf{p}\,d\mathbf{q} - \int\limits_{t_A}^{t_B} H\,dt
$$
частина яка відіграє роль функціоналу при умові фіксації повної енергії руху
$$
S_p = \int\limits_{A}^{B} \mathbf{p}\,d\mathbf{q}
$$
називається скороченною дією.

### Принцип Гюйгенса
Поступовий перехід від механічного розуміння природи розповсюдження світла до коливально-хвильового був здійснений Гюйгенсом, який показав як поняття хвильового фронта призводить до пояснення прямолінійності променів. Гамільтон надав цьому принципу строге математичне формулювання.В кожній точці середовища визначимо індикатрису швидкостей - поверхню яку описує кінець вектора швидкості світла в залежності від напрямку. В ізотропномі випадку це сфера. В анізотропному - еліпсоїд. 

![ham](./figs/fig_fronts.png)
<center><i>Локальні індикатриси швидкості в неоднорідному анізотропному середовищі<br> (з В.И. Арнольд "Математические методы классической механики").</i></center>

Границя $\Phi(t;\Phi_0)$ множини точок до яких світло доходить за час $\le t$ розповсюджуючись, починаючи від певної множини точок (точкові джерела) $\Phi_0$, називається хвильовим фронтом.

![ham](./figs/fig_huygensfront.png)
<center><i><a id='guygens'>Принцип Гюйгенса</a> <br> (з В.И. Арнольд "Математические методы классической механики"). </i></center>

Хвильовий фронт $\Phi(t+\Delta t;\Phi(t))$ є огибаючою хвильових фронтів від точок хвильового фронту $\Phi(t)$. Зрозуміло, що це фактично умова стаціонарності часу Ферма. Тому представлення процеса розповсюдження світла у вигляді хвильового фронту містить в себе променєве предcтавлення разом із принципом Ферма. Оскільки вздож фронту певна функція (час) є сталою, то фронти це  поверхні рівня цієї функції. Тоді поле градієнта цієї функції вказує напрямки розповсюдження фронту ортогональні до поверхні хвильового фронту.

![ham](./figs/fig_front1.png)
<center><i>У анізотропному середовищі напрямок розповсюдження фронту і променів може не співпадати <br> (з В.И. Арнольд "Математические методы классической механики").</i></center>

```python
ug=np.linspace(0,2*np.pi,200)
ug1=np.linspace(0.1,2*np.pi,50)
el=1.6
x = el*np.cos(ug)
y = np.sin(ug)
def front(r=0):
    fig = plt.figure()
    ax=fig.add_subplot(111)
    for t1 in ug1:
        x1 = el*np.cos(t1)+r*np.cos(ug)
        y1 = np.sin(t1)+r*np.sin(ug)
        ax.plot(x1,y1,'r',linewidth=0.2)
        ax.set_aspect(1)
    plt.plot(x,y,'y',linewidth=3)
    plt.show()
#
front(1.5)
```

### Повне внутрішнє відбиття ([total internal reflection](https://en.wikipedia.org/wiki/Total_internal_reflection))

З закона заломлення витікає, 

![theory](./figs/fig_inrefl0.png)
<center><i>Повне внутрішнє відбивання (теорія)</i></center>
що при русі з оптично щільного до менш густого середовища існує граничнй кут при якому відбувається відбивання від поверхні (заломленого проміня немає)

$$
\sin{\alpha_{\text{max}}} = \frac{n_2}{n_1}<1\,,
$$

Цей кут називається кутом Брюстера, а само явище - повним внутрішнім відбиттям.

<center><i>Повне внутрішнє відбивання (експеримент)</i></center>

Типовим проявом внутрішнього відбиття є міражі. Вони виникають завдяки температурної залежності густини і як слідство показника заломлення від температури яка змінюється з висотою над поверхнею. Більш гаряче повітря із меншим показником заломлення знаходиться нижче ніж більш холодне де показник більше.

Явище повного відбиття застосовується у оптоволоконих кабелях ([optical fibers](https://en.wikipedia.org/wiki/Optical_fiber)) - цилиндричних діелектричних середовищах в яких показник заломлення зменшується від центра до края. 

- __Задача__ Вважаючи, що при проходженні проміня центральної вісі $z$ оптоволокна кут $\alpha_0$, а показник заломлення в радіальному напрямку $r$ від вісі змінюється як $n(r) = n_0\,\sqrt{1-(r/a)^2}$, знайдіть форму проміня. 

З механічної точки зору відбиття витікає досить природньо, якщо вважати, що у оптично густому середовищі швидкість світла менша за вакуумну. Але це вочевидь протирічіть тому що нам відомо про швидкість звука. Тому можна зробити висновок що звукові та світлові колівання якісно відрізняються. Такою якістю є характеристика яка називається __поляризація__ коливань. Напрям коливань по відношенню до напряму переноса енергії може бути розкладений на продольну $\parallel$ та дві поперечні $\perp$ компоненти. Перший тип коливань називається __продольним__ і має місце як раз для довгохвильових звукових коливаннь де змішення частинок середовища відбувається у продольному напрямку (стискання-розрідження). Інший вид коливань називається __поперечним__. ЕМХ у вакуумі є прикладом поперечних коливань. У кристалічних тілах можуть виникати поперечні звукові хвилі (transverse sound).
